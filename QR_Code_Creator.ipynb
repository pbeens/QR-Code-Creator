## 2. Program & Usage
Run this cell to define the generator and create your QR code.

```python
import qrcode
from IPython.display import Image, display
import os

def create_qr_code(data, filename='qrcode.png'):
    """
    Generates a QR code from the provided data and saves it to a file.
    """
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=10,
        border=4,
    )
    qr.add_data(data)
    qr.make(fit=True)

    img = qr.make_image(fill_color="black", back_color="white")
    img.save(filename)
    print(f"QR Code saved as {filename}")
    display(Image(filename=filename))

# Interactive usage
print("Enter the URL or text for the QR Code:")
text_input = input()

if text_input:
    create_qr_code(text_input)
else:
    print("No input provided.")
```
